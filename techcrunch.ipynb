```python
melted = pd.melt(df, id_vars=['fundedDate', 'state', 'company'], value_vars=['raisedAmt'])
melted.head()
```

```python
deals = melted.set_index('fundedDate')
deals.head()
```

```python
keys = [deals.index.year]
keys
```

```python
grouper = deals.groupby(keys)
```

```python
result = grouper.sum()
result.ix[1993:2005]
```

```python
deals.head()
```

```python
grouped = deals.groupby(keys)
grouped.value.aggregate(['mean', 'count', 'sum'])
```

```python
by_year = grouped.value.aggregate(['mean', 'count', 'sum'])
by_year.columns = ['avg_dealsize', 'deals', 'volume']
by_year.head()
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(121)
ax1.plot(by_year.index, by_year.deals)


ax2 = fig.add_subplot(122)
ax2.plot(by_year.index, by_year.volume)

plt.show()
plt.clf()
```

```python
companies = df.ix[:, ['company', 'raisedAmt']]
```

```python
companies = companies.groupby('company')['raisedAmt'].sum().order(ascending=False)
```

```python
%%R
library(ggplot2)
library(plyr)
library(scales)
library(reshape2)
```

```python
company = companies.head().index
raisedAmt = companies.head().values
```

```python
%Rpush company raisedAmt
```

```python
%R company
```

```python
%%R
companies <- data.frame(company=matrix(unlist(company)), raisedAmt=matrix(unlist(raisedAmt)))
companies <- arrange(companies, desc(raisedAmt))
print(companies)
```

```python
%%R
p <- ggplot(companies, aes(reorder(company, -raisedAmt), weight=raisedAmt)) +
    geom_bar() +
    xlab('Company') +
    ylab('Total Raised') +
    scale_y_continuous(labels=dollar)

print(p)
```

```python
x_dates = by_year.index
volume = by_year.volume
```
